## plt.subplot()

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

dir_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/data/any_images'
file_names = [f for f in os.listdir(dir_path)]
plt.figure(figsize=(12, 12))

for i, file_name in enumerate(file_names):
    file_path = os.path.join(dir_path, file_name)
    image = Image.open(file_path)
    plt.subplot(1, len(file_names), i + 1)
    plt.imshow(image)
    plt.axis('off')
    plt.title(file_name)

plt.tight_layout()
plt.show()
```

## plt.subplots()

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

dir_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/data/any_images'
file_names = [f for f in os.listdir(dir_path)]

n_cols = 3
n_rows = (len(file_names) // n_cols) + (1 if len(file_names) % n_cols != 0 else 0)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 2.4, n_rows * 2.4)) # (W, H) in inches
axes = axes.flatten() # in case of more than 1 row

for i, file_name in enumerate(file_names):
    file_path = os.path.join(dir_path, file_name)
    image = Image.open(file_path)
    axes[i].imshow(image)
    axes[i].set_title(file_name)
    axes[i].axis('off')

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout() # Prevent overlap
plt.show()
```

## real color of each channel

```python
import matplotlib.pyplot as plt
import torch

img = torch.rand(size=(3, 224, 224)) # (C, H, W)
img[0, :, :] = 1.0
img[1, :, :] = 0
img[2, :, :] = 0
img_np = img.permute(1, 2, 0).numpy() # (H, W, C)

plt.figure(figsize=(12, 12))

# Display the original image
plt.subplot(1, 4, 1)
plt.title('Original')
plt.imshow(img_np)
plt.axis('off')

# Display each channel with the real color
colors = ['R', 'G', 'B']
for i in range(3):
    composite_img = torch.zeros_like(img) # Black image
    composite_img[i, ...] = img[i, ...] # Copy only each channel, leaving other channels as zero
    composite_img_np = composite_img.permute(1, 2, 0).numpy()
    plt.subplot(1, 4, i + 2)
    plt.imshow(composite_img_np)
    plt.title(f'Channel {colors[i]}')
    plt.axis('off')

# Show the plots
plt.show()
```

## gray scale of each channel

```python
import matplotlib.pyplot as plt
import torch

img = torch.zeros(size=(3, 224, 224)) # (C, H, W)
img[0, :, 112:] = 1.0  # Right-half red
img[1, :, :] = 0
img[2, :, :] = 0
img_np = img.permute(1, 2, 0).numpy() # (H, W, C)

plt.figure(figsize=(12, 12))

# Display the original image
plt.subplot(1, 4, 1)
plt.title('Original')
plt.imshow(img_np)
plt.axis('off')

# Display each channel with the real color
colors = ['R', 'G', 'B']
for i in range(3):
    channel_img = img[i, ...].numpy()  # Extract the i-th channel
    plt.subplot(1, 4, i + 2)
    plt.imshow(channel_img, cmap='gray')
    plt.title(f'Channel {colors[i]}')
    plt.axis('off')

# Show the plots
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import torch
from torchvision.io import read_image
from torchvision.transforms.functional import to_pil_image

# Define the directory path
dir_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/data/any_images'
file_names = [f for f in os.listdir(dir_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
colors = ['R', 'G', 'B']
n_cols = 4 # One column for the original image and three for the channels
n_rows = len(file_names)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 2.4, n_rows * 2.4)) # (W, H) in inches

for i, file_name in enumerate(file_names):
    file_path = os.path.join(dir_path, file_name)
    image = read_image(file_path)

    # Original image
    axes[i, 0].imshow(to_pil_image(image))
    axes[i, 0].set_title(f'Original\n{file_name}')
    axes[i, 0].axis('off')

    # Display each channel in grayscale
    for j in range(3):
        composite_img = torch.zeros_like(image) # Black image (C, H, W)
        composite_img[j, :, :] = image[j, :, :]
        composite_img_np = composite_img.permute(1, 2, 0).numpy()
        axes[i, j + 1].imshow(composite_img_np, cmap='gray')
        axes[i, j + 1].set_title(f'Channel {colors[j]}')
        axes[i, j + 1].axis('off')

plt.tight_layout()  # Prevent overlap
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import torch
from torchvision.io import read_image
from torchvision.transforms.functional import to_pil_image

# Define the directory path
dir_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/data/any_images'
file_names = [f for f in os.listdir(dir_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
colors = ['R', 'G', 'B']
n_cols = 4 # One column for the original image and three for the channels
n_rows = len(file_names)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 2.4, n_rows * 2.4)) # (W, H) in inches

for i, file_name in enumerate(file_names):
    file_path = os.path.join(dir_path, file_name)
    image = read_image(file_path)

    # Original image
    axes[i, 0].imshow(to_pil_image(image))
    axes[i, 0].set_title(f'Original\n{file_name}')
    axes[i, 0].axis('off')

    # Display each channel in grayscale
    for j in range(3):
        channel_img = image[j, :, :].numpy()  # Extract the j-th channel
        axes[i, j + 1].imshow(channel_img, cmap='gray')
        axes[i, j + 1].set_title(f'Channel {colors[j]}')
        axes[i, j + 1].axis('off')

plt.tight_layout()  # Prevent overlap
plt.show()
```
